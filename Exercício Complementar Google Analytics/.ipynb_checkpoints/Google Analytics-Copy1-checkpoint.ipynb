```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("data/dataset.csv")
```

```python
df.info()
```

```python
df.drop('Unnamed: 0', axis = 1, inplace=True)
```

```python
df.drop('fullvisitorid', axis=1, inplace=True)
```

```python
df['date'] = pd.to_datetime(df['date'],format='%Y-%m-%d')
df.info()
```

```python
# dia
df['day']=df['date'].apply(lambda x:x.day)
# mês
df['month']=df['date'].apply(lambda x:x.month)
# ano
df['year']=df['date'].apply(lambda x:x.year)
```

```python
sns.countplot(df['source_today']).set_title("Fontes de acesso ao portal")
```

**Relação entre variávies**

```python
_ = sns.lineplot(data=df, x='day', y='transactionrevenue_st', color='red', ci=None)
_.set_title('Total de faturamento por dia')
plt.show(_)
```

```python
tempo_medio_dia = df.groupby(['day'])['timeonsite_st'].mean()
tempo_medio_dia = tempo_medio_dia.reset_index(name='tempo medio')
```

```python
sns.barplot(x='day', y='tempo medio', data=tempo_medio_dia, palette='inferno')
plt.title("Média de Tempo Gasto no Portal por Dia")
```

```python
#verificando se existem colunas fortemente correlacionadas
import numpy as np
corr = df.corr()
sns.set_context("notebook", font_scale=1.0, rc={"lines.linewidth": 2.5})
plt.figure(figsize=(13,7))
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask, 1)] = True
a = sns.heatmap(corr,mask=mask, annot=True, fmt='.2f')
rotx = a.set_xticklabels(a.get_xticklabels(), rotation=90)
roty = a.set_yticklabels(a.get_yticklabels(), rotation=30)
```
