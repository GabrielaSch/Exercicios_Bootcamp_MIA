**Exercício de análise discritiva e visualização com dados reais do Google Analytics**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("data/dataset.csv")
```

# Análise Exploratória

```python
df.head()
```

```python
df.isna().any()
```

```python
df.drop('Unnamed: 0', axis = 1, inplace=True)
```

```python
df.drop('fullvisitorid', axis=1, inplace=True)
```

```python
# transformação da data para type datatime 
df['date'] = pd.to_datetime(df['date'],format='%Y-%m-%d')
df.info()
```

```python
# dia
df['day']=df['date'].apply(lambda x:x.day)
# mês
df['month']=df['date'].apply(lambda x:x.month)
# ano
df['year']=df['date'].apply(lambda x:x.year)
```

```python
plt.rcParams['figure.figsize'] = (12, 7)
```

```python
sns.countplot(df['channelgrouping']).set_title("Canais de acesso ao portal")
```

**Podemos observar que a maior parte do acesso ao portal é feita pelo canal da busca orgânica, ou seja, gratuíta. Redes sociais não é um dos canais mais usados, isso nos lembra da importância dos sites de estarem bem colocados na listagem de relevância da busca do Google**

```python
sns.countplot(df['devicecategory']).set_title("Dispositivo usado para acessar o portal")
```

**Podemos ver que a grande maioria do acesso é feita pelo desktop. Ainda que o mobile venha ganhando espaço, ainda é o computador desktop o mais utilizado.**

```python
sns.countplot(df['source_today']).set_title("Fontes de acesso ao portal")
```

**Das fontes de acesso, a maior parte é o próprio Google.**

```python
ios = pd.DataFrame(df['operatingsystem'].value_counts())
ios.plot.pie(y='operatingsystem', figsize=(10, 10),autopct='%.2f', title='Sistema Operacional usado para acessar o portal', 
                 startangle=145, legend=False, ylabel=None, 
             colors=['#ff6666', '#ffcc99', '#ffb3e7', '#99ff99', '#66b3ff', '#c2c2f0','#eaeaf2'])
```

**Como podíamos imaginar, o Sistema Operacional mais usado é o Windows.**

# Relação entre variávies

```python
_ = sns.lineplot(data=df, x='day', y='transactionrevenue_st', color='red', ci=None)
_.set_title('Total de faturamento por dia')
plt.show(_)
```

**Podemos observar três "picos" entre os dias 5 e 15, isso pode indicar uma relação com o dia do pagamento dos usuários, quando costumam efetuar mais compras online.**

```python
tempo_medio_dia = df.groupby(['day'])['timeonsite_st'].mean()
tempo_medio_dia = tempo_medio_dia.reset_index(name='tempo medio em minutos')
```

```python
sns.barplot(x='day', y='tempo medio em minutos', data=tempo_medio_dia, palette='inferno')
plt.title("Média de Tempo Gasto no Portal por Dia")
```
